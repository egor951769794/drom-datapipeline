```python
from io import BytesIO
from PIL import Image
import boto3
```

```python
print("Initializing S3 connection...")

ENDPOINT = "http://localhost:9000"
session = boto3.Session(
    aws_access_key_id="dev_admin",
    aws_secret_access_key="dev_password",
    region_name="ru-central1"
)

s3 = session.client(
    "s3",
    endpoint_url=ENDPOINT
)

print(f"S3 connection to MinIO established")
```

```python
s3r = boto3.resource(
    's3',
    endpoint_url='http://localhost:9000',
    aws_access_key_id="dev_admin",
    aws_secret_access_key="dev_password",
    region_name="ru-central1"
)
```

```python
pics_bucket = s3r.Bucket('bulletin-pictures')
pics = list(pics_bucket.objects.all())
```

```python
pic_keys = [pic.key for pic in pics]
```

```python
counter = 0
total = len(pic_keys)

for key in pic_keys:
    counter += 1
    
    response = s3.get_object(Bucket='bulletin-pictures', Key=key)
    object_content = response["Body"].read()
    
    with Image.open(BytesIO(object_content)) as pic:
        if pic.mode in ("RGBA", "P"):
            pic = pic.convert("RGB")
        w, h = pic.size
        pic = pic.resize( (int(w/2), int(h/2)), Image.LANCZOS)
    
        buffer = BytesIO()
        pic.save(buffer, format='JPEG', quality=70, optimize=True)
        buffer.seek(0)
        s3.upload_fileobj(buffer, 'bulletin-pictures', key)

        if counter % 100:
            print(f"{100* counter / total}%")
```
